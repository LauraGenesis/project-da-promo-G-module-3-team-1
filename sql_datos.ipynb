```python
#Insertar información empleado

cnx = mysql.connector.connect(
    user='root', password='AlumnaAdalab',
    host='127.0.0.1', database='abc_proyecto'
)

mycursor = cnx.cursor()


inicio = 0  
fin = 1586   


for index in tqdm(range(inicio, fin), desc="Insertando registros"):
    row = empleados.iloc[index]  
    sql = """
        INSERT INTO employee (
           Employee_id, Attrition, Age, Datebirth, Gender, Maritalstatus, Distancefromhome, Education, Educationfield
        ) VALUES (
            %s, %s, %s, %s, %s, %s, %s, %s, %s
        )
    """
   
    try:
        # Convertir valores de int64 y float64 a int y float respectivamente
        values = []
        for val in row.values:
            if isinstance(val, np.int64):
                values.append(int(val))
            elif isinstance(val, np.float64):
                values.append(float(val))
            else:
                values.append(val)
        
        mycursor.execute(sql, values)
        cnx.commit()
    except mysql.connector.Error as err:
        cnx.rollback()
        print("Error al insertar registro:", err)


mycursor.close()
cnx.close()
```
